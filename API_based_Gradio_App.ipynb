```python
%%writefile app.py
import gradio as gr
from pandasai.llm import LLM
import pandasai as pdai
from openai import OpenAI


class LLMWrapper:
    def __init__(self):
        self.client = OpenAI(
            base_url="http://localhost:8081/v1",
            api_key="fake",
        )

    def __call__(self, prompt):
        completion = self.client.chat.completions.create(
            model="fake",
            messages=[{'role': 'user', 'content': prompt}]
        )
        return completion.choices[0].message.content

llmwrapper = LLMWrapper()

class LocalLLM(LLM):
    def call(self, instruction, context):
        return llmwrapper(instruction.to_string())

    @property
    def type(self):
        return 'fake'

llm = LocalLLM()

pdai.config.set({'llm': llm})

df = None

def display_file_info(file_obj):
    if file_obj is not None:
        global df
        df = pdai.read_csv(file_obj.name)
        df_head = df.head(3)

        return df_head, gr.update(interactive=True), gr.update(interactive=True)

    else:
        return "", gr.update(interactive=False), gr.update(interactive=False)


def generate_output(user_prompt):
    if user_prompt:
        global df
        results = df.chat(user_prompt)
        print(results)
        return str(results)
    else:
        return "Prompt was empty"


with gr.Blocks(title="My PandasAI App in Colab") as demo:
    gr.Markdown("# Chat with your dataframe")

    file_input = gr.File(label="Upload your csv file here.", file_types=['.csv'])

    df_preview = gr.Dataframe(label="Dataframe Preview (First 3 Rows)", interactive=False)

    prompt_input = gr.Textbox(label="Write your command here!", lines=3, interactive=False)

    generate_button = gr.Button("Generate", interactive=False)

    generated_output = gr.Textbox(label="Answer by PandasAI", lines=20)

    file_input.upload(
        fn=display_file_info,
        inputs=file_input,
        outputs=[df_preview, prompt_input, generate_button]
    )

    generate_button.click(
        fn=generate_output,
        inputs=prompt_input,
        outputs=generated_output
    )

demo.launch(share=True)
```
